## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Remove duplicates
df = df.drop_duplicates()

# Fill missing numerical values with median
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Fill categorical with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Outlier removal using IQR
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]
```

### What all manipulations have you done and insights you found?

During the data wrangling phase, the dataset was thoroughly examined to ensure it was clean and analysis-ready. A duplicate check was performed using df.duplicated().sum(), and no duplicate records were found, confirming data integrity. Missing value analysis was conducted using df.isnull().sum(), and no explicit null values were detected across any column. Additional checks were performed to identify hidden missing indicators such as "_", "NA", or empty strings; however, none were present. Therefore, no imputation was required. Data types were verified to ensure correct classification of numerical and categorical variables. Outlier detection was carried out using distribution plots and boxplots to examine extreme values in financial attributes such as income, debt, and credit utilization. The dataset was found to be well-structured and clean, allowing the analysis to focus on extracting meaningful financial insights rather than extensive preprocessing.
